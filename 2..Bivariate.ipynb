```python
import pandas as pd
from scipy.stats import ttest_ind
from sklearn.preprocessing import LabelEncoder

dataset = dataset.dropna()

if dataset['classification'].dtype == 'object':
    le = LabelEncoder()
    dataset['classification'] = le.fit_transform(dataset['classification'])

good = dataset[dataset["appet"] == 'good']['classification']
poor = dataset[dataset["appet"] == 'poor']['classification']

result = ttest_ind(good, poor)
```

```python
result
```

```python
from scipy.stats import ttest_rel
dataset=dataset.dropna()
ckd=dataset[dataset["classification"]=='ckd']['appet']
ckd1=dataset[dataset["classification"]=='ckd']['ane']

ttest_rel(ckd,ckd1)
```
